```python
# ----------------------- EJERCICIO C.1 ----------------------- 
# Crear marco
movies.sort_values('imdb_score')
# Soluciona los problemas de asendencia.
topimbd_score=movies.sort_values('imdb_score',ascending=False)
# Filtro los datos para que imprima solo el ítulo de la película, el año, 
# el presupuesto y la puntuación imdb.
topimbd_score2=topimbd_score[
    ['movie_title','title_year','budget','imdb_score']]
```
